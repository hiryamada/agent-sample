```python
from azure.identity import AzureCliCredential
```

```python
credential = AzureCliCredential()
```

```python
from agent_framework.foundry import FoundryAgent
```

```python
import os
```

```python
# agent_name には、Foundryプロジェクトで作成・保存済みのエージェント名を指定する必要がある。
# model （デプロイ名）は指定する必要はない（指定できない）

agent = FoundryAgent(
    agent_name='agent1',
    project_endpoint=os.environ.get('PROJECT_ENDPOINT'),
    credential=credential
)
```

```python
response = await agent.run("東京の今日の天気は？")
```

```python
print(response)
```

```python
from IPython.display import Markdown, display
```

```python
display(Markdown(response.text))
```

```python
# セッションは自動的には作成されない。

response = await agent.run("私の名前は山田です")
display(Markdown(response.text))

response = await agent.run("私の名前は?")
display(Markdown(response.text))
```

```python
# セッションを明示的に作成

session = agent.create_session()

response = await agent.run("私の名前は山田です", session=session)
display(Markdown(response.text))

response = await agent.run("私の名前は?", session=session)
display(Markdown(response.text))
```
